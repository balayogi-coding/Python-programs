```python
# Declaring List
# List can contain any number of values.
# List can have different kinds of data types.
# list counting start with 0
emptyList = []
emptyList = list()

myList = [1,2.5,'hello',True,1]
print(myList)

# To print specific value
print(myList[1])
```

```python
# To append a value to the list
myList.append('Image')
print(myList)
```

```python
# To insert into specific position
myList.insert(2,'Five')
print(myList)
```

```python
# To find Index of the element
myList.index(2.5)
```

```python
# To pop element from the list
popped_element = myList.pop()
print(popped_element)
```

```python
# To count the elements in the list
newlist = [1,2,2,3,4,5,2,5,6]
newlist.count(2)
```

```python
# To reverse the elements in the list
newlist.reverse()
newlist
```

```python
# To extend the list at the end
myList.extend(newlist)
myList
```
